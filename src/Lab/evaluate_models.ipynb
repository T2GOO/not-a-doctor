```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import os
from os.path import dirname, isfile, join
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import h5py
```

```python
root = dirname(os.getcwd())

val_dir = os.path.join(root, 'data', 'current', 'test')
IMAGE_DIM = 150
BATCH = 32
```

```python
path_to_models =join(root, 'model')
model_paths = [f for f in os.listdir(path_to_models) if isfile(os.path.join(path_to_models, f))]
print(model_paths)
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)

test_data = test_datagen.flow_from_directory(
    val_dir,
    target_size=(IMAGE_DIM, IMAGE_DIM),
    batch_size=BATCH,
    color_mode = 'rgb',
    class_mode='categorical',
    shuffle = False
)

losses = []
accuracies = []
nb_param = []
```

```python
def evaluate_model(model_path, generator):
    model = load_model(model_path)
    total_params = model.count_params()
    loss, accuracy = model.evaluate(generator, verbose=1)
    return loss, accuracy, total_params
```

```python
# Need to go to the model directory to open .h5 file, very nice...
os.chdir(path_to_models)
for path in model_paths:
    print(f"\nModele evaluation: {path}")
    try:
        loss, accuracy, total_params = evaluate_model(path, test_data)
        losses.append(loss)
        accuracies.append(accuracy)
        nb_param.append(total_params)
        print(f"Loss: {loss:.4f}, Accuracy: {accuracy:.4f}")
    except Exception as e:
        print(e)
        losses.append(10e10)
        accuracies.append(0)
        nb_param.append(0)


print("\nPerformance benchmark :")
for i, path in enumerate(model_paths):
    print(f"{path} - Perte: {losses[i]:.4f}, Précision: {accuracies[i]:.4f}")

plt.figure(figsize=(10, 6))
plt.bar(range(len(model_paths)), accuracies, tick_label=model_paths)
plt.ylabel("Accuracy")
plt.title("Performance benchmark")
plt.xticks(rotation=45)
plt.show()
```

Looking for a balance between the number of parameters and the accuracy/loss.
Create a score to define the best model (few parameters with high accuracy)

```python
modele_scores = []
for i, path in enumerate(model_paths):
    score = accuracies[i] / (max(nb_param[i], 1) * losses[i]) * 100_000_000
    modele_scores.append(score)
    print(f"{path} - score: {score}")
            
```

```python
import heapq

top3_scores = heapq.nlargest(3, modele_scores)
top3_modeles = []
top3_modeles = [model_paths[modele_scores.index(score)] for score in top3_scores]
print(top3_modeles)
```
